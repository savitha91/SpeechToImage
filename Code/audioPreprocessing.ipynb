## STEPS to extract audio features from .mp4 file
First we analyse the time-domain, frequency-domain and spectogram data from the audio.
Using windowing function (Hanning) and striding , we get to know the number of time-frames (1 to t). 
t is calculated using (inputShape - windowSize)/ stride +1. This is similar to the calculation used in CNN.
On Applying RFFT(Fourier transform) on the samples at time-frames, we get the amplitude values at respective frequencies.
This gives array of shape [windowSize/2 , (inputShape - windowSize)/ stride +1], where the columns formula same as t(mentioned above). row is windowSize/2, since, we consider one-dimensional discrete Fourier Transform for real input(rfft)

We can infer that x1, x2 ,..... xt , will have windowSize/2 amplitude values, which will act as input to the Speech recognition model

## SPECTOGRAM

Optimum window size is 20-30ms,as humans cant utter more than one phoneme in this span.
For Speech-recognition, we consider window overlap(stride) as 50%

```python

def spectrogram(samples, sample_rate, stride_ms = 10.0, 
                          window_ms = 20.0, max_freq = None, eps = 1e-14):
    print("Number of data pointes(samples) per second", sample_rate)
    print("Duration of audio", len(samples)/sample_rate)
    stride_size = int(0.001 * sample_rate * stride_ms)
    window_size = int(0.001 * sample_rate * window_ms)
    print("Len of samples",len(samples))
    print("stride size",stride_size)
    print("window_size",window_size)
    # Extract strided windows
    truncate_size = (len(samples) - window_size) % stride_size
    print("truncate_size",truncate_size)
    samples = samples[:len(samples) - truncate_size]
    print("samples",samples)
    # In image ,we calculate shape of the output image(after passign through conv2D) using input shape(nHI), stride and filter shapes,
    # as nH = (nHI - f / s) + 1. Same is done here. (inputSize-windowSize / strideSize)+1 = 103 
    # shape of output data is (960,103)
    nshape = (window_size, (len(samples) - window_size) // stride_size + 1)
    print("nshape",nshape)
    nstrides = (samples.strides[0], samples.strides[0] * stride_size) #This sample has 4 strides
    print("nstrides",nstrides)
    windows = np.lib.stride_tricks.as_strided(samples, 
                                          shape = nshape, strides = nstrides)
    # windows : gives the values of the windows i.e 960*103 values
    
    assert np.all(windows[:, 1] == samples[stride_size:(stride_size + window_size)])

    # Window weighting, squared Fast Fourier Transform (fft), scaling
  
    weighting = np.hanning(window_size)[:, None] # creates tapers using weighted consine, for window size 960. Each weight is a measure of amplitude
    print("weighting",len(weighting)) 
    plt.title("After applying Hanning window function on raw data")
    plt.xlabel("(time)No of datapoints in 1 window [Here, 960 points]")
    plt.ylabel("Amplitude in m")
    plt.plot(np.arange(window_size), windows) #1 window graph. 960
    #plt.xlim(0, 10)
    #plt.xlim(0, window_size-1)
    #plt.ylim(-0.025, 1.025)
    # From the plot we can see that hanning function touches zero
    
    # Here the windows/amplitudes have real values(not complex values),hence we use rfft(r->real value)\
    val = windows * weighting
    # rfft applied on amplitudes in each window of size 960 
    fft = np.fft.rfft(val, axis=0) # Computes fft for positive part of the wave
    print("fft.shape",fft.shape)
    fft = np.absolute(fft)
    
    freqs = np.fft.rfftfreq(len(val), d=1/window_size)
    print(len(freqs))
    
    plt.figure(figsize=(7,10))
    plt.title("FT applied on the wave created using Hanning function. Amplitude measured in m")
    #This gives the amplitudes at  frequencies from 0 to 480Hz
    plt.xlabel("Frequencies in 1 window")
    plt.ylabel("Amplitude in m")
    plt.plot(freqs, fft) # Frequencies in the window of shape(960)
    #plt.xlim(0, )
    #plt.xlim(0, 20)
    
    plt.figure(figsize=(7,10))
    plt.title("FT applied on the wave created using Hanning function. Amplitude measured in Decibel")
    plt.xlabel("Frequencies in 1 window")
    plt.ylabel("Amplitude in db")
    plt.plot(freqs, 20*np.log10(fft / fft.max()))
    #plt.xlim(0, )
    fft = fft**2 #squaring the fft values
    
    
    scale = np.sum(weighting**2) * sample_rate
    print("scale",scale)
    fft[1:-1, :] *= (2.0 / scale)
    fft[(0, -1), :] /= scale
    
    # Prepare fft frequency list
    freqs = float(sample_rate) / window_size * np.arange(fft.shape[0])
    #print("freqs",freqs)
    # Compute spectrogram feature
    ind = np.where(freqs <= max_freq)[0][-1] + 1
    print("ind",ind)
    specgram = np.log(fft[:ind, :] + eps)
    return specgram
```

```python
spec = spectrogram(samples, sample_rate, stride_ms = 10.0, 
                          window_ms = 20.0, max_freq = sample_rate/2)
print("Shape of specs array", spec.shape)
```

### Shape of specs array  = (481, 103)
Number of time-steps(t) after applying window_size and stride_size is 103 . 
Calculated using (inputShape - windowSize)/ strideSize +1. inputshape = Total number of sampling points(= 49920) in the entire dataset considering sampling_rate as 48KHz

Every time-step(t) has frequencies ranging from 0 to 481. spec array has amplitude values at these frequencies

This will become the input X = {x1,x2,x3, ....... , x<103>}, where, x1 has 481 amplitude values(samples)
X will be fed to the model

we can consider this as image classification prob, once we have the spec array
